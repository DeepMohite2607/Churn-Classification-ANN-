```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
```

```python
# load the pickle Files
model = load_model('model.h5')

with open('standard_scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('onehot_encoder_geo.pkl', 'rb') as f:
    encoder = pickle.load(f)

with open('label_encoder_gender.pkl', 'rb') as f:
    label_encoder = pickle.load(f)
```

```python
#Example input data for prediction
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encoded = encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=encoder.get_feature_names_out(['Geography']))
```

```python
input_data = pd.DataFrame([input_data])
```

```python
input_data = pd.concat([input_data.reset_index(drop=True),geo_encoded_df],axis=1)
input_data.head()
```

```python
#Encode categorical variables
input_data['Gender'] = label_encoder.transform(input_data['Gender'])
```

```python
## Concatentate the encoded columns with the original DataFrame
input_df = input_data.drop('Geography', axis=1)
```

```python
input_df.head()
```

```python
input_scaled = scaler.transform(input_df)
```

```python
##Prediction Churn 

predictions = model.predict(input_scaled)

print(predictions)
```

```python
predictions_proba = predictions[0][0]
```

```python
predictions_proba
```

```python
if predictions_proba > 0.5:
    print("The customer is likely to churn.")
else:   
    print("The customer is not likely to churn.")
```
